```python
import pandas as pd
import numpy as np
```

```python
ppl=pd.read_csv('people.csv')
act=pd.read_csv('act_train.csv')
```

```python
ppl['date']=ppl['date'].apply(lambda x: pd.to_datetime(x,format="%Y-%m-%d") )
act['date']=act['date'].apply(lambda x: pd.to_datetime(x,format="%Y-%m-%d") )
```

```python
act2=pd.merge(act,ppl,how='left',left_index=True,on='people_id',suffixes=('_act','_ppl'))
act2.columns
```

```python
for cc in act2.columns:
    if act2[cc].dtype=='bool':
        act2[cc]=act2[cc]*1
    if "char" in cc:
        act2[cc]=act2[cc].fillna('type 0')
```
